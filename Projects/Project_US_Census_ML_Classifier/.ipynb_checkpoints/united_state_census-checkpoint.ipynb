## Data Understanding and Cleaning

```python
import numpy as np # importing the numpy librarie.
import pandas as pd # importing the pandas librarie.
import matplotlib.pyplot as plt # importing the matplotlib.pyplot librarie.
import seaborn as sns # importing the seaborn librarie.
LABELS = [">50K", "<=50k"] # Labels for our dataset.
```

```python
data = pd.read_csv('census.csv') # Reading the Csv file and storing it as pandas dataframe.
data.columns # Columns in the dataset.
```

```python
print(data.shape) # Printing number of rows and columns in the dataset.
data.head() # Printing top 5 rows in the dataset.
```

```python
data.describe(include='all') # Performing Statistical analysis on the dataset.
```

Let us now check the missing values in the dataset

```python
data.isnull().values.any() # Checking for missing values in the dataset.
```

```python
print((data['occupation']=="?").any()) # Checking for miscoded values in the 'occupation' column.
print((data['native.country']=="?").any()) # Checking for miscoded values in the 'native.country' column.
print((data['workclass']=="?").any()) # Checking for miscoded values in the 'workclass' column.
```

### Observation: 
Here, we notice that some of the values in the dataset are miscoded for the columns "occupation", "native.country" and "workclass".
So, we are replacing those values with the most occuring value in the respective colummn.

```python
data['occupation']=data['occupation'].str.replace('?','Prof-specialty') # replacing the miscode values in the 'occupation' column
print((data['occupation']=="?").any())                                  # with the most occuring value in that column.
```

```python
data['native.country']=data['native.country'].str.replace('?','United-States') # replacing the miscode values in the 'native.country' 
print((data['native.country']=="?").any())                                # column with the most occuring value in that column.
```

```python
data['workclass']=data['workclass'].str.replace('?','Private') # replacing the miscode values in the 'workclass' column 
print((data['workclass']=="?").any())                          # with the most occuring value in that column.
```

## Data Prepration

```python
features = list(set(data.columns)-set(data['income'])) # Extracting list of all other features except 'income'.
a = data[features] # Storing all columns as input for modelling the data.
b = data['income'] # Storing the 'income' column as output for modelling the data.
```

```python
X = pd.get_dummies(a) # Converting all the input columns to numerical columns.
print(X.columns) # Printing the columns after converting to numeric data.
```

```python
from sklearn.model_selection import train_test_split # Importing sklearn module for test-train split.
x_train, x_test, y_train, y_test = train_test_split(X, b, test_size=0.3, random_state=0 ) # Storing all the test train data.
```

***
## Model Building and Evaluation

### 1. Logistic Regression


```python
from sklearn.linear_model import LogisticRegression # Importing logistic Regression package from sklearn model.
regressor = LogisticRegression() # Creating the regressor object.
regressor.fit(x_train,y_train) # Fitting the training data. 
```

```python
y_pred = regressor.predict(x_test) # Predicting the output column for training data.
```

### Accuracy Score

```python
from sklearn.metrics import accuracy_score # Importing the accuracy_score package from sklearn module.
s1 = accuracy_score(y_test,y_pred) # Computing the accuracy score for the model.
print("Accuracy(%age):",s1*100) # Printing the accuracy score.
```

### 2. K-Nearest Neighbours



```python
from sklearn.neighbors import KNeighborsClassifier # Importing KNeighborsClassifier package from sklearn module.
classifier = KNeighborsClassifier() # Creating the classifier object.
classifier.fit(x_train, y_train) # Fitting the training data.
```

```python
y_pred_1 = classifier.predict(x_test)  # Predicting the output column for training data.
```

### Accuracy Score

```python
from sklearn.metrics import accuracy_score # Importing the accuracy_score package from sklearn module.
s2 = accuracy_score(y_test,y_pred_1) # Computing the accuracy score for the model.
print("Accuracy(%age):",s2*100) # Printing the accuracy score.
```

### 3. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier # Importing RandomForestClassifier package from sklearn module.
gini_clf1 = RandomForestClassifier(criterion='gini') # Creating the classifier object.
gini_clf1.fit(x_train,y_train) # Fitting the training data.
```

```python
y_pred_2 = gini_clf1.predict(x_test) # Predicting the output column for training data.
```

### Accuracy Score

```python
from sklearn.metrics import accuracy_score # Importing the accuracy_score package from sklearn module.
s3 = accuracy_score(y_test, y_pred_2) # Computing the accuracy score for the model.
print("Accuracy(%age):",s3*100) # Printing the accuracy score.
```

### Best Model Evaluation

```python
accuracy = [s1,s2,s3] # Storing all the accuracy scores in an array.
data1 = {'accuracy_score':accuracy, # Making a dictionary for accuracy score and algorithms used.
        'Algorithm':["Logistic Regression","K-Nearest Neighbours","Random Forest"]}
df = pd.DataFrame(data1) # Converting dictionary to pandas dataframe.
sns.set(rc={'figure.figsize':(8,6)}) # Setting figure size for seaborn graph.
sns.barplot(x=df.Algorithm, y = df.accuracy_score) # Plotting barplot for evaluating all the algorithms.
plt.show()
```

## Observation:
So from all the three algorithms applied **Random Forest** stands out to give the best results.
