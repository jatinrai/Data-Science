```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('weather_3.csv')
df.head()
```

```python
df.dropna()
```

```python
sns.distplot(df.Rainfall,bins=10)
plt.show()
```

```python
sns.distplot(df.Evaporation,hist=True,kde=False)
plt.show()
```

```python
sns.distplot(df.Sunshine,kde=True,hist=False)
plt.show()
```

```python
sns.kdeplot(df.MinTemp,shade=True)
sns.kdeplot(df.MaxTemp,shade=True)
plt.show()
```

```python
df.plot(kind='scatter',x='MaxTemp',y='Sunshine')
plt.show()
```

```python
sns.jointplot(x='MinTemp',y='Rainfall',data=df)
plt.show()
```

```python
df.columns
```

```python
with sns.axes_style('darkgrid'):
    sns.jointplot(x='MinTemp',y='Evaporation',data=df,kind='hex',color='k')
plt.show()
```

```python
with sns.axes_style('darkgrid'):
    sns.jointplot(x='MaxTemp',y='Sunshine',data=df,kind='hex',color='k')
plt.show()
```

```python
sns.jointplot(x='MaxTemp',y='Sunshine',data=df,kind='kde')
plt.show()
```

```python
sns.jointplot(x='MinTemp',y='Evaporation',data=df,kind='kde')
plt.show()
```

```python
data = df[['MinTemp','MaxTemp','Rainfall','Evaporation']]
sns.pairplot(data)
plt.show()
```

```python
sns.stripplot(x='WindDir9am',y='MaxTemp',data=df)
plt.show()
```

```python
sns.swarmplot(x='WindGustDir',y='MaxTemp',data=df)
plt.show()
```

```python
sns.stripplot(x='WindDir9am',y='MaxTemp',data=df,jitter=True)
plt.show()
```

```python
sns.boxplot(x = 'Cloud9am', y = 'MinTemp', data= df)
plt.show()
```

```python
sns.boxplot(x = 'WindGustDir', y = 'MinTemp', data= df)
plt.show()
```

```python
sns.barplot(x='RainTomorrow', y='Rainfall', data=df)
plt.show()
```

```python
sns.barplot(x='RainToday', y='Rainfall', data=df)
plt.show()
```

```python
sns.barplot(x='Cloud3pm', y='Humidity3pm', data=df, hue='RainToday')
plt.show()
```

```python
sns.countplot(x='WindGustDir', data=df)
plt.show()
```
